# Klasifikasi Penentuan Penerima Bantuan PKH

# Data Preparation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Dataset/DataTrain.xlsx')
data.head(10)
```

# Data Selection

```python
target = 'KETERANGAN'
```

```python
datalatih=data[['USIA','JUMLAH_TANGGUNGAN','PEKERJAAN','PENGHASILAN','TEMPAT_TINGGAL', 'JENIS_LANTAI', 'JENIS_DINDING', 'BANTUAN_LAIN','KETERANGAN']]
```

```python
x_train = datalatih.drop(["KETERANGAN"], axis = 1)
x_train.head()
```

```python
y_train = datalatih["KETERANGAN"]
y_train.head(10)
```

```python
datalatih.head(10)
```

```python
pd.options.mode.chained_assignment = None
```

```python
datalatih.isna().sum()
```

# Classification

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

modelnb = GaussianNB()
nbtrain = modelnb.fit(x_train, y_train)
```

```python
data2 = pd.read_excel('Dataset/DataTest.xlsx')
data.head(10)
```

```python
datauji=data2[['USIA','JUMLAH_TANGGUNGAN','PEKERJAAN','PENGHASILAN','TEMPAT_TINGGAL', 'JENIS_LANTAI', 'JENIS_DINDING', 'BANTUAN_LAIN','KETERANGAN']]
```

```python
pd.crosstab(datauji['JUMLAH_TANGGUNGAN'], datauji['KETERANGAN'])
```

```python
x_test = datauji.drop(["KETERANGAN"], axis = 1)
x_test.head()
x_test.isna().sum()
```

```python
y_test = datauji["KETERANGAN"]
y_test.head(10)
```

```python
pd.options.mode.chained_assignment = None
```

```python
datalatih.isna().sum()
```

```python
y_pred = nbtrain.predict(x_test)
y_pred
```

```python
np.array(y_test)
```

```python
nbtrain.predict_proba(x_test)        
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

# Classification Report

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test, y_pred)
print("Akurasi Naive Bayes : ",accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
